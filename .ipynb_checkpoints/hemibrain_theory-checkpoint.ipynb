```python
import numpy as np
import torch

def estimate_cov_eigs_linear(T_sim, dt_save, dt, W, sigma_noise=1,
                             T_save_delay=100, N_batch=1, N_loops=10, return_vecs=False):

    device = W.device
    N = W.shape[0]
    eval_iter = int(dt_save / dt)
    Nt = int(T_sim / dt)
    N_save = int(T_sim / dt_save)
    # define stuff
    x_save = torch.zeros(N_save, N_batch, N, device=device)
    x = torch.randn(N_batch, N, device=device) * torch.std(W).item() * np.sqrt(N)
    x_save[0] = x
    # run
    x_xcovs = torch.zeros(N_loops, N_batch, N, N, device=device)
    for i_loop in range(N_loops):
        print('loop:{}'.format(i_loop))
        x_save = torch.zeros(N_save, N_batch, N, device=device)
        x = torch.randn(N_batch, N, device=device) * torch.std(W).item() * np.sqrt(N)
        x_save[0] = x
        for i in range(1, Nt):
            noise = torch.randn(N_batch, N, device=device)
            x += dt * (-x + torch.mm(x, W.T) + sigma_noise*noise)
            if i % eval_iter == 0:
                x_save[i // eval_iter] = x
        x_save = x_save[int(T_save_delay / dt_save):]
        x_xcov = torch.einsum('tbi, tbj -> bij', x_save, x_save) / (x_save.shape[0] - 1)
        x_xcovs[i_loop] = x_xcov
    x_xcov_mean = x_xcovs.mean(dim=(0, 1))

    return x_xcov_mean

def calculate_evals(x_xcov_mean):
    print('calculating eigenvalues')
    eigs_x, vecs_x = torch.linalg.eigh(x_xcov_mean)
    if not return_vecs:
        return eigs_x.cpu().detach().numpy()
    else:
        return (eigs_x.cpu().detach().numpy(),
                vecs_x.cpu().detach().numpy())


```

```python
U = torch.from_numpy(U_sk).type(torch.FloatTensor).to(0)
V = torch.from_numpy(V_sk).type(torch.FloatTensor).to(0)
S = torch.from_numpy(s_sk).type(torch.FloatTensor).to(0)
gamma_diag = torch.einsum('ik, ik -> k', U, V)
N = len(S)
```

```python
### --- Generate LDRG matrix --- ###

X1 = torch.randn((N, N)).to(0)
X2 = torch.randn((N, N)).to(0)
Y = torch.randn((N, N)).to(0)
rho = gamma_diag
L = torch.sqrt((1 - torch.abs(rho))/N)*X1 + torch.sqrt(torch.abs(rho)/N)*Y
R = torch.sqrt((1 - torch.abs(rho))/N)*X2 + torch.sign(rho)*torch.sqrt(torch.abs(rho)/N)*Y

#g_eff = 8
D = torch.clone(S)
#g_correction = g_eff / torch.sqrt(torch.sum(D**2)/N)
#D = D * g_correction

W = torch.einsum('ik, k, jk -> ij', L, D, R)
del L
del R
```
