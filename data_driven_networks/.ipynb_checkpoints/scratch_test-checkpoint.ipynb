```python
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# Nonstationary *periodic* GP template + phase-shifted population
# + softplus-like x->r map (your formula, normalized per neuron)
#
# Kernel:
#   K(,') = ()(') exp(-2 sin^2((u()-u('))/2)/^2) + _n^2 I
#   u() =  + a1 sin  + a2 sin 2  (phase warp -> nonstationary around cycle)
#   () = 0 (1 + v cos(-v))      (phase-dependent amplitude)
# ------------------------------------------------------------

def warped_phase(theta, a1=0.35, a2=0.15):
    return theta + a1*np.sin(theta) + a2*np.sin(2*theta)

def amp_mod(theta, sigma0=1.0, v=0.45, theta_v=1.1):
    s = sigma0*(1 + v*np.cos(theta - theta_v))
    return np.clip(s, 1e-6, None)

def periodic_se_kernel(u, ell=0.6):
    du = u[:, None] - u[None, :]
    return np.exp(-2.0*(np.sin(du/2.0)**2) / (ell**2))

def nonstationary_periodic_kernel(theta, sigma0=1.0, v=0.45, theta_v=1.1,
                                  a1=0.35, a2=0.15, ell=0.6, sigma_n=0.05):
    u = warped_phase(theta, a1=a1, a2=a2)
    base = periodic_se_kernel(u, ell=ell)
    s = amp_mod(theta, sigma0=sigma0, v=v, theta_v=theta_v)
    K = (s[:, None]*s[None, :]) * base
    K = K + (sigma_n**2)*np.eye(len(theta))
    return K

def chol_from_kernel(theta, kernel_kwargs):
    K = nonstationary_periodic_kernel(theta, **kernel_kwargs)
    jitter = 1e-8*np.eye(len(theta))
    return np.linalg.cholesky(K + jitter)

def sample_from_chol(L, rng, n_samples=1):
    z = rng.standard_normal((L.shape[0], n_samples))
    return (L @ z).T  # (n_samples, T)

def shift_by_phase(f, theta, mu_i):
    th = (theta - mu_i) % (2*np.pi)
    theta_wrap = np.r_[theta, 2*np.pi]
    f_wrap = np.r_[f, f[0]]
    return np.interp(th, theta_wrap, f_wrap)

def softplus_rate(x, beta=6.0, b=0.25, target_mean=1.0, axis=-1):
    # r = (1/Z) log(1 + exp(beta(x-b))), choose Z to hit target_mean over 
    y = np.log1p(np.exp(beta*(x - b)))
    #Z = np.mean(y, axis=axis, keepdims=True) / max(target_mean, 1e-9)
    return y #/ Z

# -----------------------------
# Generate synthetic peri-gait population
# -----------------------------
rng = np.random.default_rng(0)

T = 160
theta = np.linspace(0, 2*np.pi, T, endpoint=False)

# Tune these to taste
kernel_main = dict(
    sigma0=1.0, v=0.45, theta_v=1.0,
    a1=0.35, a2=0.15,
    ell=0.75,
    sigma_n=0.03
)

kernel_eps = dict(
    sigma0=0.18, v=0.0, theta_v=0.0,
    a1=0.0, a2=0.0,
    ell=0.35,
    sigma_n=0.02
)

L_main = chol_from_kernel(theta, kernel_main)
L_eps  = chol_from_kernel(theta, kernel_eps)

# Shared nonstationary template current x0()
x0 = sample_from_chol(L_main, rng, n_samples=1)[0]
x0 = x0 - np.mean(x0)

# Population: phase-shifted copies + neuron-specific texture
N = 2000
mu = np.linspace(0, 2*np.pi, N, endpoint=False)      # phase offsets (tessellate cycle)
A  = np.exp(rng.normal(0.0, 0.25, size=N))           # per-neuron amplitude variation

eps = sample_from_chol(L_eps, rng, n_samples=N)      # (N, T) small texture

X = np.zeros((N, T))
for i in range(N):
    X[i] = A[i]*shift_by_phase(x0, theta, mu[i]) + eps[i]

# Map currents -> rates via your softplus-like formula
beta = 1
b = -1
R = softplus_rate(X, beta=beta, b=b, target_mean=1.0, axis=1)  # normalize per neuron

# Sort neurons by peak phase (creates diagonal ridge like the figure)
peak_idx = np.argmax(R, axis=1)
order = np.argsort(peak_idx)
R_sorted = R[order]

# -----------------------------
# Plot
# -----------------------------
plt.figure(figsize=(10, 4))
plt.imshow(R_sorted, aspect='auto', origin='lower',
           extent=[0, 2*np.pi, 0, N])
plt.xlabel(r'Gait phase $\theta$')
plt.ylabel('Neuron (sorted by peak)')
plt.title('Synthetic peri-gait rates: nonstationary periodic GP template + phase shifts')
plt.colorbar(label='rate (a.u.)')
plt.tight_layout()
plt.show()

mid = order[N//2]
plt.figure(figsize=(10, 3))
plt.plot(theta, x0, label='template current $x_0(\\theta)$')
plt.plot(theta, X[mid], label='example neuron current $x_i(\\theta)$', alpha=0.8)
plt.xlabel(r'$\theta$')
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 3))
plt.plot(theta, R[mid], label='example neuron rate $r_i(\\theta)$')
plt.xlabel(r'$\theta$')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np

def periodic_central_diff(X, theta):
    """
    Periodic central difference dX/dtheta for X shape (N,T) on uniform theta grid.
    """
    dtheta = float(theta[1] - theta[0])
    return (np.roll(X, -1, axis=1) - np.roll(X, 1, axis=1)) / (2.0 * dtheta)

def fit_J_from_orbit(X, R, theta, ridge=1e-2, fit_bias=True):
    """
    Fit J (and optional bias b) in the phase-domain regression:

        Xdot + X    J @ R  (+ b)

    where:
      X, R : (N, T) arrays over phase grid theta
      theta: (T,) uniform grid over [0, 2)
      ridge: L2 regularization strength
      fit_bias: if True, solve Xdot+X  J@R + b[:,None]

    Returns:
      J : (N,N)
      b : (N,) or None
      diag : dict with mse, r2
    """
    N, T = X.shape
    if R.shape != (N, T):
        raise ValueError("R must have same shape as X (N,T).")
    if theta.shape != (T,):
        raise ValueError("theta must have shape (T,).")

    Xdot = periodic_central_diff(X, theta)
    Y = Xdot + X                          # target (N,T)

    if fit_bias:
        # Augment regressors with a constant row to fit b
        R_aug = np.vstack([R, np.ones((1, T))])          # (N+1, T)
        G = R_aug @ R_aug.T                              # (N+1, N+1)
        W = (Y @ R_aug.T) @ np.linalg.inv(G + ridge*np.eye(N+1))  # (N, N+1)
        J = W[:, :N]
        b = W[:, N]
        Yhat = J @ R + b[:, None]
    else:
        G = R @ R.T                                      # (N, N)
        J = (Y @ R.T) @ np.linalg.inv(G + ridge*np.eye(N))
        b = None
        Yhat = J @ R

    resid = Y - Yhat
    mse = float(np.mean(resid**2))
    denom = float(np.sum((Y - Y.mean(axis=1, keepdims=True))**2) + 1e-12)
    r2 = float(1.0 - np.sum(resid**2) / denom)

    return J, b, {"mse": mse, "r2": r2, "ridge": float(ridge), "fit_bias": bool(fit_bias)}

# ---- Use it like this (assuming you already have X, R, theta) ----
# J_hat, b_hat, diag = fit_J_from_orbit(X, R, theta, ridge=1e-2, fit_bias=True)
# print(diag)

# a little least squares as a treat: scan ridge
# for lam in [0.0, 1e-4, 1e-2, 1e-1, 1.0]:
#     J, b, d = fit_J_from_orbit(X, R, theta, ridge=lam, fit_bias=True)
#     print(lam, d)
```

```python
def fit_J_with_global_inhibition(X, R, theta, ridge=1e-2):
    """
    Fit J and h in:
        Xdot + X  J @ R + h[:,None] * m
    where m(t) = population mean of R at each phase.
    """
    N, T = X.shape
    if R.shape != (N, T):
        raise ValueError("R must have shape (N,T)")
    if theta.shape != (T,):
        raise ValueError("theta must have shape (T,)")

    # dX/d + X
    dtheta = float(theta[1] - theta[0])
    Xdot = (np.roll(X, -1, axis=1) - np.roll(X, 1, axis=1)) / (2.0 * dtheta)
    Y = Xdot + X

    # population-mean regressor m()
    m = R.mean(axis=0, keepdims=True)          # (1, T)

    # augmented regressors [R; m]
    R_aug = np.vstack([R, m])                  # (N+1, T)

    G = R_aug @ R_aug.T                        # (N+1, N+1)
    W = (Y @ R_aug.T) @ np.linalg.inv(G + ridge*np.eye(N+1))  # (N, N+1)

    J = W[:, :N]                               # recurrent weights
    h = W[:, N]                                # global-inhibition coefficients per neuron

    Yhat = J @ R + h[:, None] * m
    resid = Y - Yhat
    mse = float(np.mean(resid**2))
    denom = float(np.sum((Y - Y.mean(axis=1, keepdims=True))**2) + 1e-12)
    r2 = float(1.0 - np.sum(resid**2) / denom)

    return J, h, {"mse": mse, "r2": r2, "ridge": float(ridge)}
```

```python
import numpy as np

# ---------------------------------------------
# 1. Fit J from the peri-gait orbit
# ---------------------------------------------
def periodic_central_diff(X, theta):
    """
    Periodic central difference dX/dtheta for X shape (N,T) on uniform theta grid.
    """
    dtheta = float(theta[1] - theta[0])
    return (np.roll(X, -1, axis=1) - np.roll(X, 1, axis=1)) / (2.0 * dtheta)

def fit_J_from_orbit(X, R, theta, ridge=1e-2, fit_bias=True):
    """
    Fit J (and optional bias b) in:
        Xdot + X    J @ R  (+ b)

    X, R : (N, T)
    theta: (T,)
    """
    N, T = X.shape
    if R.shape != (N, T):
        raise ValueError("R must have shape (N,T).")
    if theta.shape != (T,):
        raise ValueError("theta must have shape (T,)")

    Xdot = periodic_central_diff(X, theta)
    Y = Xdot + X  # (N,T)

    if fit_bias:
        R_aug = np.vstack([R, np.ones((1, T))])          # (N+1, T)
        G = R_aug @ R_aug.T                              # (N+1, N+1)
        W = (Y @ R_aug.T) @ np.linalg.inv(G + ridge*np.eye(N+1))  # (N, N+1)
        J = W[:, :N]
        b = W[:, N]
        Yhat = J @ R + b[:, None]
    else:
        G = R @ R.T                                      # (N, N)
        J = (Y @ R.T) @ np.linalg.inv(G + ridge*np.eye(N))
        b = None
        Yhat = J @ R

    resid = Y - Yhat
    mse = float(np.mean(resid**2))
    denom = float(np.sum((Y - Y.mean(axis=1, keepdims=True))**2) + 1e-12)
    r2 = float(1.0 - np.sum(resid**2) / denom)
    return J, b, {"mse": mse, "r2": r2, "ridge": float(ridge), "fit_bias": bool(fit_bias)}

# ---------------------------------------------
# 2. Use underdetermined directions to enforce transverse stability
# ---------------------------------------------
def stabilize_transverse(J, R, rank=None, alpha=1.0, shrink_offdiag=False, gamma=0.0, seed=0):
    """
    Construct J_stab that preserves J's action on the orbit subspace spanned by R
    as much as possible, while enforcing strong contraction on the orthogonal complement.

    Steps:
      1. Find an orthonormal basis U for the column-space ("orbit subspace") of R.
      2. Complete {U} to an orthonormal basis Q = [U, U_perp].
      3. Transform J into this basis: J_tilde = Q^T J Q.
      4. Leave blocks that affect the orbit (J11, J21) unchanged.
      5. Replace purely transverse block J22 with -alpha * I.
         Optionally shrink J12 (transverse -> orbit) by factor gamma.
      6. Transform back: J_stab = Q J_tilde_mod Q^T.

    Args
    ----
    J : (N,N) array, fitted effective connectivity
    R : (N,T) array of firing rates along the orbit
    rank : int or None
        Number of singular vectors for the orbit subspace.
        If None, picked by >= 99% energy of singular values.
    alpha : float
        Contraction strength in transverse directions (>0).
    shrink_offdiag : bool
        If True, shrink J12 (transverse -> orbit) by factor gamma.
    gamma : float
        Multiplicative factor for J12 if shrink_offdiag=True. gamma=0  zero that block.
    seed : int
        Random seed for constructing the orthogonal complement.

    Returns
    -------
    J_stab : (N,N) stabilized connectivity
    info   : dict with diagnostic info, including how much JR changed
    """
    N, T = R.shape
    if J.shape != (N, N):
        raise ValueError("J must be (N,N)")

    # Orbit subspace via SVD of R
    U_svd, S, _ = np.linalg.svd(R, full_matrices=False)
    if rank is None:
        energy = np.cumsum(S**2) / np.sum(S**2)
        k = int(np.searchsorted(energy, 0.99)) + 1
    else:
        k = int(rank)
    k = min(k, N)
    U = U_svd[:, :k]  # (N,k)

    # Complete to orthonormal basis Q = [U, U_perp]
    rng = np.random.default_rng(seed)
    if k == N:
        Q = U
        k_perp = 0
    else:
        B = rng.standard_normal((N, N-k))
        M = np.hstack([U, B])  # (N,N)
        Q, _ = np.linalg.qr(M)
        k_perp = N - k

    # Transform J into this basis
    J_tilde = Q.T @ J @ Q
    J11 = J_tilde[:k, :k]
    J12 = J_tilde[:k, k:]
    J21 = J_tilde[k:, :k]
    J22 = J_tilde[k:, k:]

    # Modify transverse block
    if k_perp > 0:
        J22_new = -alpha * np.eye(k_perp)
        if shrink_offdiag:
            J12_new = gamma * J12
        else:
            J12_new = J12

        top = np.hstack([J11, J12_new])
        bottom = np.hstack([J21, J22_new])
        J_tilde_mod = np.vstack([top, bottom])
    else:
        J_tilde_mod = J_tilde

    # Transform back to original coordinates
    J_stab = Q @ J_tilde_mod @ Q.T

    # Diagnostics: how much does J R change along the orbit?
    JR = J @ R
    JR_stab = J_stab @ R
    diff = JR_stab - JR
    rel_err = np.linalg.norm(diff) / (np.linalg.norm(JR) + 1e-12)
    info = {
        "k_orbit": k,
        "alpha": float(alpha),
        "shrink_offdiag": bool(shrink_offdiag),
        "gamma": float(gamma),
        "rel_change_JR": float(rel_err),
    }
    return J_stab, info

R = phi_numpy(X_filtered)
J, bias, res = fit_J_from_orbit(X_filtered, R, theta, ridge=1e-2, fit_bias=False)
J_stab, info = stabilize_transverse(J, R)
```
